```python
plt.barh(df["Make"], df["Fastcharge Speed"],color=["lightblue", "lightgrey"])
plt.title("Fastcharge Speed")
plt.xlabel("Fastcharge Speed")
plt.ylabel("Make")
plt.show()
```

```python
plt.barh(df["Make"], df["Electric Range"], color=["lightblue", "lightgrey"])
plt.title("Electric Range per model")
plt.xlabel("Electric Range")
plt.ylabel("Make")
plt.show()
```

```python
plt.barh(df["Make"], df["Combined - Mild Weather"],color=["lightblue", "lightgrey"])
plt.title("Performance on combined terrain and mild weather")
plt.xlabel("Make")
plt.ylabel("Combined - Mild Weather")
plt.grid(color="whitesmoke", linestyle="--", linewidth=0.9)
plt.show()
```
